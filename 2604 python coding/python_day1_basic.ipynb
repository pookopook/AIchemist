# 1차수: 변수와 기본 연산자

## 1. 파이썬 코드 실행
- 파이썬은 위에서 아래로 실행됩니다
- 한 줄씩 순서대로 실행됩니다

```
print("첫 번째")
print("두 번째")
print("세 번째")
```

## 2. 변수
- 변수는 값을 저장하는 공간입니다

```
a = 10
b = 20

print(a)
print(b)
```

### 💡 실습
값을 바꿔보세요

```
a = 5
b = 7

print(a + b)
```

## 3. 자료형
- 정수(int)
- 실수(float)
- 문자열(str)

```
num = 10
pi = 3.14
text = "hello"

print(num, pi, text)
```

## 4. 산술 연산자

```
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```
