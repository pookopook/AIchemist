```python
# gpt 버전
## 피라미드 두개(위, 아래) 각각 찍어서 합치기
m = 5
for i in range(1, m):
    print(' ' *(m-i-1) + '*' * (i*2-1))
for i in range(m-2, 0, -1):
    print(' ' *(m-i-1) + '*' * (i*2-1))
```

```python
# for문을 한번만 쓰는 버전
## 중앙을 중심으로 늘어났다 줄어들도록 하기
m = 10
mid = m // 2
for i in range(1, m):
    k = mid - abs(mid - i)  # abs 절댓값
    print(' ' * (mid - k) + '*' * (k * 2 - 1))
```

## BMI 계산기
 - BMI를 계산하는 방법
     - BMI 공식: 체중 / 키(m)^2
     - 비만위험도: 정상- ~23 미만, 위험-~25미만, 비만- 25이상~  
<p>
 - 요구사항
     - 키와 체중을 입력받고,
     - BMI 수치와 비위험도를 출력하세요.  

    * (참고) 대한비만학회 비만진료지침 2022 개정 8판

```python
키 = float(input('키(cm) 입력') or 175.0)
체중 = float(input('체중(kg) 입력') or 80.0)

bmi = 체중 / (키/100)**2
print(f'BMI - {bmi}, {키}, {체중}')

if bmi < 23:
    print('정상')
elif bmi < 25:
    print('위험')
else:
    print('비만')
```
