## Q. 다이아몬드 출력하기

```
예시
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```

````python
# gpt 버전
## 피라미드 두개(위, 아래) 각각 찍어서 합치기
m = 5
for i in range(1, m):
    print(' ' *(m-i-1) + '*' * (i*2-1))
for i in range(m-2, 0, -1):
    print(' ' *(m-i-1) + '*' * (i*2-1))
````

````python
# for문을 한번만 쓰는 버전
## 중앙을 중심으로 늘어났다 줄어들도록 하기
m = 10
mid = m // 2
for i in range(1, m):
    k = mid - abs(mid - i)  # abs 절댓값
    print(' ' * (mid - k) + '*' * (k * 2 - 1))
````

## BMI 계산기
 - BMI를 계산하는 방법
     - BMI 공식: 체중 / 키^2
     - 비만위험도: 정상- \~22.9, 위험- \~24.9, 비만- 25\~  
<p>
 - 요구사항
     - 키와 체중을 입력받고,
     - BMI 수치와 비위험도를 출력하세요.  

    * (참고) 대한비만학회 비만진료지침 2022 개정 8판

````python
키 = input('키(cm)를 입력하세요.') or 175 # or를 쓸 경우 기본값이 된다.
체중 = input('체중(kg)을 입력하세요.') or 75
키, 체중 = int(키)/100, int(체중)

bmi = 체중 / 키**2
print(f'BMI는 {bmi}입니다.')
print(type(bmi), type(키))
if   bmi <  23.0:
    print(' 정상입니다.')
elif bmi <= 24.9:
    print(' 비만전단계입니다.')
else:
    print('비만위험 비만입니다.')
````

````python
# 구현 1
# 계정이 1개인 세상
email = input('이메일 입력하세요') or 'test@mirinae.com'

if '@' in email and '.' in email:
    pw = input('암호를 입력하세요') or 'abc123'
    if email == 'test@mirinae.com':
        if pw == 'abc123':
            print('로그인되었습니다.')
        else:
            print('암호가 틀렸습니다.')
else:
    print('이메일 형식이 맞지 않습니다.')
````

## 가위바위보
 - 가위바위보 게임 만들기
 - 유저는 가위, 바위, 보 중 하나를 입력
 - 컴퓨터는 자동선택
 - 각자 선택지와 게임 승패를 출력

````python
# 기본버전
import random
a = input('게임시작 가위, 바위, 보!') or '가위'
b = random.choice(['가위','바위','보'])

if a not in ['가위','바위','보']:
    print('가위바위보 중 하나를 입력하세요')

else:
    print(f'유저:{a}, 컴퓨터:{b}')
    if a == b:
        print('비겼습니다.')
    elif (a == '가위' and b == '보') or \
         (a == '바위' and b == '가위') or \
         (a == '보'   and b == '바위'):       # 승리조건을 일일히 매칭해줘야 함
        print('이겼습니다!')
    else:
        print('졌습니다.')
````

````python
# 모듈러 방식으로 승패 판정
import random
mapping = {'가위':0,'바위':1,'보':2}  # 딕셔너리

a = input('게임시작 가위, 바위, 보!') or '가위'
b = random.choice(['가위','바위','보'])

if a not in ['가위','바위','보']:
    print('가위바위보 중 하나를 입력하세요')
else:
    print(f'유저:{a}:{mapping[a]}, 컴퓨터:{b},{mapping[b]}')
    mod = (mapping[a] - mapping[b]) % 3
    if mod == 0:
        print('비겼습니다.')
    elif mod == 1:
        print('이겼습니다!')
    else:
        print('졌습니다.')
````
