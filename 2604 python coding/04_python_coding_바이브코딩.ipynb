## 가위바위보
 - 가위바위보 게임 만들기
 - 유저는 가위, 바위, 보 중 하나를 입력
 - 컴퓨터는 자동선택
 - 각자 선택지와 게임 승패를 출력

```python
# 기본버전
import random
a = input('게임시작 가위, 바위, 보!') or '가위'  # or '가위' 는 테스트용
b = random.choice(['가위','바위','보'])

if a not in ['가위','바위','보']:
    print('가위바위보 중 하나를 입력하세요')

else:
    print(f'유저:{a}, 컴퓨터:{b}')
    if a == b:
        print('비겼습니다.')
    elif (a == '가위' and b == '보') or \
         (a == '바위' and b == '가위') or \
         (a == '보'   and b == '바위'):       # 승리조건을 일일히 매칭해줘야 함
        print('이겼습니다!')
    else:
        print('졌습니다.')
```

```python
import random

choices = ["가위", "바위", "보"]

print("가위바위보 게임")
print("종료하려면 '종료'를 입력하세요.")

while True:  # 무한 반복
    user = input("가위, 바위, 보 중 하나를 입력하세요: ")

    if user == "종료":
        print("게임을 종료합니다.")
        break

    if user not in ["가위", "바위", "보"]:
        print("잘못 입력했습니다. 다시 입력하세요.")
        continue

    computer = random.choice(choices)

    print(f"사용자: {user}")
    print(f"컴퓨터: {computer}")

    # 텍스트 기반
    if user == computer:
        print("비겼습니다.")
    elif (
        (user == "가위" and computer == "보") or
        (user == "바위" and computer == "가위") or
        (user == "보" and computer == "바위")
    ):
        print("이겼습니다.")
    else:
        print("졌습니다.")

    # 모듈러 연산
    result = (user - computer) % 3

    if result == 0:
        print("비겼습니다.")
    elif result == 1:
        print("이겼습니다.")
    else:
        print("졌습니다.")
        
```

```python
import random

choices = ["가위", "바위", "보"]

print("가위바위보 게임")
print("0: 가위, 1: 바위, 2: 보")
print("종료하려면 q를 입력하세요.")

while True:
    user_input = input("선택하세요(0, 1, 2): ")

    if user_input == "q":
        print("게임을 종료합니다.")
        break

    if user_input not in ["0", "1", "2"]:
        print("잘못 입력했습니다.")
        continue

    user = int(user_input)
    computer = random.randint(0, 2)

    print(f"사용자: {choices[user]}")
    print(f"컴퓨터: {choices[computer]}")

    result = (user - computer) % 3

    if result == 0:
        print("비겼습니다.")
    elif result == 1:
        print("이겼습니다.")
    else:
        print("졌습니다.")
```

```python
# 모듈러 방식으로 승패 판정
import random
mapping = {'가위':0,'바위':1,'보':2}  # 딕셔너리

a = input('게임시작 가위, 바위, 보!') or '가위'
b = random.choice(['가위','바위','보'])

if a not in ['가위','바위','보']:
    print('가위바위보 중 하나를 입력하세요')
else:
    print(f'유저:{a}:{mapping[a]}, 컴퓨터:{b},{mapping[b]}')
    mod = (mapping[a] - mapping[b]) % 3
    if mod == 0:
        print('비겼습니다.')
    elif mod == 1:
        print('이겼습니다!')
    else:
        print('졌습니다.')
```

```python
# 작업 규칙

## 1. 목적
이 문서는 이 저장소에서 작업할 때 반복적으로 지켜야 하는 운영 원칙을 정리한 문서입니다.

핵심 목표는 세 가지입니다.
- raw 데이터 보존
- `final_case_id` 중심의 일관된 연결 유지
- 단계별 산출물의 재현 가능성 확보

## 2. 데이터 흐름 기본 원칙
- raw는 가능한 한 수정하지 않습니다.
- 사람이 손으로 고쳐야 하는 판단은 review 단계에서 수행합니다.
- review 결과는 다음 단계 master/mart에 반영되도록 연결합니다.
- downstream 결과가 이상할 때는 raw를 직접 고치기보다, 어떤 단계에서 값이 바뀌었는지 추적합니다.

## 3. final_case_id 우선 원칙
- downstream 단계에서는 `final_case_id`를 가장 중요한 공통 키로 봅니다.
- 제조, 수입, inventory, mart, 모델링, 보고서는 모두 `final_case_id`를 기준으로 연결합니다.
- `inventory_case_id`나 원천 `case_id`는 보조 키로만 사용합니다.
- 새 테이블이나 새 산출물을 만들 때도 가능하면 `final_case_id`를 포함시킵니다.

## 4. raw 보존 원칙
- raw 폴더의 파일은 원본 자산으로 간주합니다.
- raw의 값이 이상해 보여도, 바로 원본 파일을 수정하는 방식은 피합니다.
- 먼저 inventory, parse review, case-id review, normalize, alias 사전 같은 보정 레이어에서 해결 가능한지 검토합니다.
- raw 수정이 꼭 필요하면 별도 기록을 남깁니다.

## 5. review와 apply 원칙
- review는 사람이 판단하는 단계이고, apply는 그 판단을 데이터셋에 반영하는 단계입니다.
- review 산출물은 가능한 한 재사용 가능한 csv/xlsx 형태로 남깁니다.
- clipboard 입력을 쓰더라도, 결과적으로는 다시 읽을 수 있는 테이블 형태가 남아야 합니다.
- 자동 규칙과 사람 판단이 섞이는 경우, 어떤 값이 자동인지 수동인지 구분 가능해야 합니다.

## 6. 파일 형식 원칙
- 사람 검토용 산출물은 `xlsx`, `csv`, `pdf` 같은 눈으로 확인하기 쉬운 형식을 우선합니다.
- 파이프라인 내부 재사용용 산출물은 `pkl` 또는 `parquet` 같은 기계 친화 형식을 우선합니다.
- 동일한 내용을 사람용과 시스템용으로 둘 다 보관할 필요가 있으면 `csv/xlsx + pkl` 조합을 허용합니다.
- 이미지, 차트, SVG 같은 시각 자료는 보고서나 review를 돕는 보조 자산으로 취급합니다.

## 7. 노트북과 src 역할 분리
- 노트북은 실행 순서와 운영 절차를 보여주는 진입점입니다.
- 실제 핵심 로직은 가능한 한 `src/` 아래 함수로 둡니다.
- 노트북 셀에 큰 비즈니스 로직을 오래 남겨두지 않습니다.
- 재사용 가능성이 있는 로직은 `src.costing` 또는 관련 모듈로 내립니다.

## 8. 단계별 책임 분리
- `0-X`: 작업 설정과 환경 준비
- `1-X`: inventory와 원천 파일 정리
- `2-X`: parse, review, `final_case_id` 확정, master 정리
- `3-X`: 분석 mart, 모델링, 통합 점수, 보고서
- `4-X`: 템플릿 산출물 생성

특히 아래 구분을 유지합니다.
- `3-2`: feature mart 생성
- `3-2A`: 재료 피어 검토 보조 단계
- `3-3`: 모델링 및 통합 점수
- `3-4`: PDF 보고서 생성

## 9. 라벨링과 보정 규칙
- 재료 라벨링은 상시 업무가 아니라, 원가산정 회차 기준의 오프라인 보정 작업으로 봅니다.
- 모든 재료를 전수 라벨링하지 않습니다.
- 비교군 구성에 영향을 많이 주는 후보만 우선 검토합니다.
- 라벨은 이력으로 누적하되, 실제 반영은 최신 승인 결과를 우선합니다.
- 공급자 정보는 강한 힌트이지만, 단독으로 동일 재료를 의미하지는 않습니다.

## 10. 공정 데이터 해석 규칙
- 공정 데이터는 단순 테이블이 아니라 원본 행 순서가 의미를 갖는 블록형 데이터일 수 있습니다.
- 이런 구조에서는 함부로 정렬하지 않습니다.
- 공정명 누락 보정은 원본 순서를 보존한 상태에서 처리합니다.
- 상세행 시간은 실제 공정시간으로 인정하되, 외주가공 `Y`인 경우는 시간 집계에서 제외합니다.
- 가공품 연결이 있으면 그 연결을 우선하고, 없으면 공정연번 순차 연결을 fallback으로 사용합니다.

## 11. 문서 관리 원칙
- 구조 설명, 규칙, 설계 의도는 `docs/`에 남깁니다.
- 한 번 결정된 운영 규칙은 대화에만 두지 말고 문서로 승격합니다.
- 한글 문서는 반드시 UTF-8로 저장하고, 깨진 파일은 부분 수정이 아니라 내용 검증 후 복구합니다.
- 노트북 본문은 한글 깨짐 위험이 있으면 영어로 유지하고, 설명은 md 문서로 보완합니다.

## 12. 변경 시 체크할 것
새 로직이나 새 산출물을 추가할 때는 최소한 아래를 확인합니다.
1. 이 데이터는 어느 단계의 책임인가?
2. 공통 키는 무엇인가?
3. 사람 검토가 필요한가, 자동 처리 가능한가?
4. 재실행 시 같은 결과를 다시 만들 수 있는가?
5. review 산출물과 system 산출물을 구분했는가?
6. 관련 문서를 같이 갱신했는가?

## 13. 권장 작업 방식
- 먼저 현재 단계의 입력과 출력 파일을 확인합니다.
- 그다음 노트북보다 `src/` 함수를 먼저 읽어 실제 책임 위치를 파악합니다.
- 규칙성 있는 판단은 함수/사전/테이블로 옮깁니다.
- 사람이 판단해야 하는 부분은 review 워크플로로 분리합니다.
- 변경 후에는 다음 단계가 기대하는 입력 형식이 유지되는지 확인합니다.
```
