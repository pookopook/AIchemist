```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt

lat = 37.3404
lon = 127.9485

G = ox.graph_from_point((lat, lon), dist=1500, network_type='drive')

# ❌ 이 줄은 지운다
# G_simplified = ox.simplify_graph(G)

# 바로 무방향 그래프로 변환
G_u = nx.Graph(G)

pos = {n: (d['x'], d['y']) for n, d in G_u.nodes(data=True)}
```

```python
G_u = nx.Graph(G)  # 무방향 그래프
pos = {n: (d['x'], d['y']) for n, d in G_u.nodes(data=True)}
```

```python
# 1. Degree centrality 계산
deg_c = nx.degree_centrality(G_u)   # dict {node: value}

# 2. 색상/크기용 리스트 만들기
values = list(deg_c.values())
node_size = [v * 800 for v in values]   # 배수는 취향껏 조절
node_color = values                     # 값에 따라 colormap 적용

plt.figure(figsize=(8, 8))
nx.draw(
    G_u, pos,
    node_size=node_size,
    node_color=node_color,
    cmap='viridis',
    edge_color='lightgray',
    linewidths=0
)
plt.title("Degree Centrality - Bangok-dong Road Network")
plt.axis('off')
plt.show()
```

```python
# 1. 노드 betweenness centrality
btw_c = nx.betweenness_centrality(G_u, normalized=True)

values = list(btw_c.values())
node_size = [v * 3000 for v in values]  # betweenness는 값이 작아서 크게 곱해줌
node_color = values

plt.figure(figsize=(8, 8))
nx.draw(
    G_u, pos,
    node_size=node_size,
    node_color=node_color,
    cmap='plasma',
    edge_color='lightgray',
    linewidths=0.3
)
plt.title("Betweenness Centrality (Potentially Busy Intersections)")
plt.axis('off')
plt.show()
```

```python
edge_btw = nx.edge_betweenness_centrality(G_u, normalized=True)

# 엣지 색상 리스트
edge_colors = [edge_btw[e] for e in G_u.edges()]

plt.figure(figsize=(8, 8))
nx.draw(
    G_u, pos,
    node_size=0,
    edge_color=edge_colors,
    edge_cmap=plt.cm.inferno,
    width=2
)
plt.title("Edge Betweenness (Potentially Busy Road Segments)")
plt.axis('off')
plt.show()
```

```python
import random

# 1. 노드 두 개 랜덤 선택
nodes = list(G_u.nodes())
src = random.choice(nodes)
dst = random.choice(nodes)

# 2. 최단경로 (가중치: edge length)
# OSMnx 그래프에는 기본으로 'length'(미터) 속성이 있음
path = nx.shortest_path(G_u, source=src, target=dst, weight='length')

print("최단 경로 노드 수:", len(path))

# 3. 경로를 이루는 엣지 목록 만들기
path_edges = list(zip(path[:-1], path[1:]))

plt.figure(figsize=(8, 8))

# 전체 도로는 회색으로
nx.draw(
    G_u, pos,
    node_size=0,
    edge_color='lightgray',
    width=1
)

# 최단경로만 색/두께 강조
nx.draw_networkx_nodes(
    G_u, pos,
    nodelist=path,
    node_size=30,
    node_color='red'
)
nx.draw_networkx_edges(
    G_u, pos,
    edgelist=path_edges,
    width=3,
    edge_color='red'
)
plt.title("Shortest Path Example")
plt.axis('off')
plt.show()
```

```python
from networkx.algorithms.community import greedy_modularity_communities

# 1. 커뮤니티 계산
communities = list(greedy_modularity_communities(G_u))

# 노드 → 커뮤니티 인덱스 매핑
comm_idx = {}
for i, comm in enumerate(communities):
    for n in comm:
        comm_idx[n] = i

node_color = [comm_idx[n] for n in G_u.nodes()]

plt.figure(figsize=(8, 8))
nx.draw(
    G_u, pos,
    node_size=5,
    node_color=node_color,
    cmap='tab20',
    edge_color='lightgray',
    width=0.5
)
plt.title("Communities in Bangok-dong Road Network")
plt.axis('off')
plt.show()

print("커뮤니티 수:", len(communities))
```
