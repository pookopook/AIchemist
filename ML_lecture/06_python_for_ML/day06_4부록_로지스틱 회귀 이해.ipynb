```python
import numpy as np
import matplotlib.pyplot as plt

# 시그모이드 함수 정의
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# x 값 범위 설정 (독립 변수)
# -10부터 10까지 100개의 점 생성
x = np.linspace(-10, 10, 100)

# 회귀 계수 (예시 값)
# beta_0 (절편) = 0, beta_1 (기울기) = 1
beta_0 = 0
beta_1 = 1

# 선형 조합 z = beta_0 + beta_1 * x
z = beta_0 + beta_1 * x

# 시그모이드 함수를 적용하여 확률 (y) 계산
y_prob = sigmoid(z)

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(x, y_prob, label=r'$\frac{1}{1 + e^{-(0 + 1x)}}$', color='blue')
plt.axhline(0.5, color='red', linestyle='--', label='y=0.5 (결정 경계)')
plt.axvline(0, color='gray', linestyle=':') # z=0이 되는 x값
plt.title('Logistic Regression Sigmoid Curve')
plt.xlabel('x (독립 변수)')
plt.ylabel('P(Y=1|x) (확률)')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```
