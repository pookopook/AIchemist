```python
df = pd.read_excel('250124 엑셀 연습 파일.xlsx',
             sheet_name = '9셀병합',
             skiprows=4 , usecols=['구분', '부서', '상품코드', '상품명', '단가(원)', '판매수량(개)', '총액(원)'])

first_row = df['구분'].str.contains('병합을') == True
first_index = df[first_row].index

df = df.iloc[:first_index[0], :]
df = df[df['상품코드'].notna()]

df[['구분', '부서']] = df[['구분', '부서']].ffill()

df
```
